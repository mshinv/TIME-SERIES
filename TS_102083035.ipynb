```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dropout, Dense, BatchNormalization, Flatten, Conv1D, LSTM
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler
```

```python
df= pd.read_excel(r"C:\Users\PRAVEEN\Downloads\DATASET.xlsx")
```

```python
df.head()
```

```python
df=df.drop(['Unnamed: 0'], axis=1)
```

```python
data_x=[]
data_y=[]
```

```python
for i in range(len(df)):
    if (df['year'][i]==9 and df['year'][i+1]!= 10) or df['year'][i]==10:
        continue
    data_x.append(df[i:i+1])
    data_y.append(df[i+1:i+2])
```

```python
for i in range(len(data_x)):
    data_x[i]= data_x[i].to_numpy()
    data_y[i]= data_y[i].to_numpy()
```

```python
deletion_rows=[]
for i in range(len(data_x)):
    if np.isnan(np.min(data_x[i])):
        deletion_rows.append(i)
print(deletion_rows)
```

```python
for i in deletion_rows:
    data_x.pop(i)
    data_y.pop(i)
```

```python
def get_model3():
    model = Sequential()
    model.add(LSTM(100, input_shape=(1,14), return_sequences=True))
    model.add(Dense(14))
    model.compile(loss='mse', optimizer='Adam', metrics=[[tf.keras.metrics.RootMeanSquaredError()]])
    return model
```

```python
data_x= np.asarray(data_x)
data_y= np.asarray(data_y, dtype='float32')
```

```python
model= get_model3()
model.fit(data_x, data_y, epochs=100, batch_size=128)
```

```python
def get_predictions(x):
    my_pred=[]
    start_index=x
    start_index=np.expand_dims(start_index,axis=1)
    for i in range(3):
        sample=np.asarray(start_index)
        start_index=model.predict(sample)
        my_pred.append(start_index)

    print('prediction for 10th year\n')
    print(my_pred[0])
    print('prediction for 11th year\n')
    print(my_pred[1])
    print('prediction for 12th year\n')
    print(my_pred[2])
```
